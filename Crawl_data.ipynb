```python
import pandas as pd
import yfinance as yf
yf.pdr_override()
```

```python
#Lấy dữ liệu về cổ phiếu của NVIDIA corporation từ ngày 22/1/1999 đến 5/1/2024
stock_name = 'NVDA'
start = '1999-01-22' 
end = '2024-01-05'
df = yf.download(stock_name, start, end)
df = df.reset_index()
```
